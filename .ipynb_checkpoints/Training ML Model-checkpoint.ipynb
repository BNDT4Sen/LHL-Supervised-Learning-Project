## Random Forest

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score
```

```python
X = diabetes_DF.drop(['Outcome'], axis = 1)
y = diabetes_DF['Outcome']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 999)
```

```python
Estimator_Count = 200

RFC = RandomForestClassifier(n_estimators = Estimator_Count, random_state=0)

RFC.fit(X_train, y_train)

y_pred = RFC.predict(X_test)

print(f'Model accuracy score with {Estimator_Count} decision-trees : {accuracy_score(y_test, y_pred)}')
print(f'Model recall score with {Estimator_Count} decision-trees : {recall_score(y_test, y_pred)}')
print(f'Model precision score with {Estimator_Count} decision-trees : {precision_score(y_test, y_pred)}')
print(f'Model F1 score with {Estimator_Count} decision-trees : {f1_score(y_test, y_pred)}')
```

```python
feature_scores = pd.Series(RFC.feature_importances_, index=X_train.columns).sort_values(ascending=False)

feature_scores
```

```python
plt.legend([],[], frameon=False)

plt.xlabel('Feature Importance Score')

plt.ylabel('Features')

plt.title("Visualizing Important Features")

sns.barplot(
    x=feature_scores, 
    y = feature_scores.index, 
    order = feature_scores.index[::-1], 
    hue = feature_scores, 
    palette = 'rocket', 
    alpha = 0.9, 
    width = 0.7, 
    dodge = False
)

plt.legend([],[], frameon=False)
print()
```

Now the random forest will be remade without the three least important features.

```python
X = diabetes_DF.drop(['Outcome', 'SkinThickness', 'Pregnancies', 'Insulin'], axis = 1)
y = diabetes_DF['Outcome']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 999)
```

```python
Estimator_Count = 200

RFC2 = RandomForestClassifier(n_estimators = Estimator_Count, random_state=0)

RFC2.fit(X_train, y_train)

y_pred = RFC2.predict(X_test)

print(f'Model accuracy score with {Estimator_Count} decision-trees : {accuracy_score(y_test, y_pred)}')
print(f'Model recall score with {Estimator_Count} decision-trees : {recall_score(y_test, y_pred)}')
print(f'Model precision score with {Estimator_Count} decision-trees : {precision_score(y_test, y_pred)}')
print(f'Model F1 score with {Estimator_Count} decision-trees : {f1_score(y_test, y_pred)}')
```

## Random Forest Summary
The model accuracy for the second random forest is marginally higher than the first. Compared to the Principal Component Analysis, Pregnancies were rated as being dramatically less impactful on the incidence of diabetes. More testing would be needed to determine what exactly causes the discrepancies between the two models.

The F1 score is unimpressive, with the recall score being especially awful. This makes the model poor for its intended purpose - identifying individuals with diabetes with minimal false negatives.
